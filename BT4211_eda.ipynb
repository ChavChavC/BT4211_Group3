### Analysis on Continuous Variables

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Avg_Price'], kde=True)
plt.title('Distribution of Avg_Price')
plt.xlabel('Avg_Price')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Avg_Price'])
plt.title('Box Plot of Avg_Price')
plt.xlabel('Avg_Price')
plt.show()
```

```python
# Count occurrences of each value in the 'Quantity' column
quantity_counts = df['Quantity'].value_counts()

# Print the counts
print("Counts of each occurrence in the 'Quantity' column:")
print(quantity_counts)
```

```python
numerical_columns = ['Tenure_Months','Avg_Price', 'Online_Spend']
max_values = {'Avg_Price': 1400}

for col in numerical_columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')

    # Set y-axis limits based on maximum values
    if col in max_values:
        plt.ylim(0, max_values[col] + 100)

    plt.show()
```

```python
import seaborn as sns

product_category_counts = df['Product_Category'].value_counts()

top_11_categories = product_category_counts.head(11).index

df_top_11 = df[df['Product_Category'].isin(top_11_categories)]

plt.figure(figsize=(10, 6))
sns.countplot(data=df_top_11, x='Product_Category', order=top_11_categories, palette='viridis')
plt.title('Distribution of Top 11 Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])
df['Quarter'] = df['Transaction_Date'].dt.to_period('Q')
```

```python
latest_record_date = df['Transaction_Date'].max()
earliest_record_date = df['Transaction_Date'].min()
print("Latest record date:", latest_record_date)
print("Earliest record date:", earliest_record_date)
```

```python
category_sales = df.groupby('Product_Category')['Avg_Price'].sum()

total_sales = df['Avg_Price'].sum()
category_percentage = (category_sales / total_sales) * 100

category_sales_df = pd.DataFrame({'Total_Sales': category_sales, 'Percentage_of_Total_Sales': category_percentage})

category_sales_df['Rank'] = category_sales_df['Percentage_of_Total_Sales'].rank(ascending=False)


category_sales_df = category_sales_df.sort_values(by='Rank')

print("Ranked product categories based on their contribution to total sales:")
print(category_sales_df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))

plt.pie(category_sales_df['Percentage_of_Total_Sales'], labels=category_sales_df.index, autopct='%1.1f%%', startangle=140)

plt.title('Contribution of Product Categories to Total Sales')

plt.axis('equal')  

plt.show()
```

```python
purchase_categories = ['Nest-USA', 'Nest', 'Nest-Canada', "Apparel", "Bags", "Drinkware", "Office"]
filtered_df = df[df['Product_Category'].isin(purchase_categories)]

average_tenure_months = filtered_df.groupby('Product_Category')['Tenure_Months'].mean()

print("Average tenure months for purchases under the specified categories:")
print(average_tenure_months)
```

## Understanding customer purchase patterns

```python
customer_purchase_counts = df['CustomerID'].value_counts()

customers_with_multiple_purchases = customer_purchase_counts[customer_purchase_counts > 1]

if len(customers_with_multiple_purchases) > 0:
    print("Customers who have made multiple purchases:")
    print(customers_with_multiple_purchases)
```

```python
# Calculate counts of each category in Coupon_Status
coupon_counts = df['Coupon_Status'].value_counts()

# Calculate percentage of each category
coupon_percentage = coupon_counts / len(df) * 100

# Combine counts and percentages into a DataFrame
coupon_summary = pd.DataFrame({'Count': coupon_counts, 'Percentage': coupon_percentage})

# Print the summary
print(coupon_summary)
```

```python
import matplotlib.pyplot as plt

# Plot counts and percentages
plt.figure(figsize=(10, 6))

# Plot combined graph
ax = coupon_counts.plot(kind='bar', color='skyblue')
coupon_percentage.plot(kind='bar', color='lightgreen', alpha=0.5)  # Overlay percentage bars

plt.title('Coupon Status Counts and Percentages')
plt.xlabel('Coupon Status')
plt.ylabel('Count / Percentage')
plt.xticks(rotation=45)

for bar1, bar2 in zip(ax.patches, coupon_percentage):
    height1 = bar1.get_height()
    height2 = bar2 / 100 * height1  # Calculate the height for percentage display
    plt.text(bar1.get_x() + bar1.get_width() / 2, height1, f'{height1}', ha='center', va='bottom')
    plt.text(bar1.get_x() + bar1.get_width() / 2, height1 - 5000, f'{bar2:.2f}%', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
